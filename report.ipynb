### Q1 - PCA IMPLEMENTATION



Principal Component Analysis (PCA) is implemented by first centering the data around its mean and then calculating the covariance matrix. Next, eigenvectors and eigenvalues are computed from the covariance matrix. Finally, these eigenvectors are sorted, and a subset is chosen to represent the principal components, reducing the dimensionality of the data.

### Q2 - PROBABILISTIC PCA IMPLEMENTATION


In probabilistic PCA, a probabilistic model is used to infer latent variables from observed data. It assumes a Gaussian distribution for the latent variables and introduces noise. The model maximizes the likelihood of the observed data given the latent variables, allowing for dimensionality reduction while accounting for uncertainty in the data.

# QUESTION 3 - VARIATIONAL AUTO ENCODERS

Variational Autoencoders (VAEs) employ a neural network architecture comprising an encoder and decoder. The encoder maps input data to a probability distribution in a latent space, incorporating mean and variance. The decoder reconstructs data from samples drawn from this distribution. VAEs optimize a variational lower bound to learn efficient representations while generating data.

1) Effect of Latent Variables on MSE:
      As the number of latent variables increased, the MSE consistently decreased across all models, indicating improved reconstruction quality with higher-dimensional latent spaces.


2) Comparative Analysis of MSE Values:
  
  Order of MSE Values: PCA > PPCA > VAE
      
  Similarity between PCA and PPCA: PCA and PPCA exhibited comparable MSE values, suggesting a similarity in their ability to capture underlying patterns in the data.


3) Performance difference with VAE:
      VAE outperformed both PCA and PPCA in terms of MSE, demonstrating superior reconstruction accuracy. This underscores the effectiveness of probabilistic and variational approaches in capturing complex relationships within the data.

Our findings highlight the influence of latent variable dimensionality on image reconstruction quality. While PCA and PPCA showed similarity, the probabilistic nature of PPCA contributed to a nuanced difference. VAE, employing a variational approach, demonstrated superior performance, emphasizing its capacity to model intricate data structures.
